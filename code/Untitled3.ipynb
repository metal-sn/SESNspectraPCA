```python
fsum = [noise[i:i+2].sum() for i in range(0,100,2)]
fsum = np.asarray(fsum)
```

```python
wv = np.arange(10,1000,20)
```

```python
fsum.shape
```

```python
plt.plot(w, noise, label="Original resolution")
plt.plot(wv, fsum,label="1/2 resolution")
plt.xlabel('Wavelength', fontsize=14)
plt.ylabel('Flux', fontsize=14)
plt.title('Fake Spectrum', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```

```python
plt.plot(w, noise, 'o', markersize=3, label="Original resolution")
plt.plot(wv, fsum, 'o', markersize=3, label="1/2 resolution")
plt.xlabel('Wavelength', fontsize=14)
plt.ylabel('Flux', fontsize=14)
plt.title('Fake Spectrum', fontsize=16)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
