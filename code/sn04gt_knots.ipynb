```python
sn04gt = "test_template.txt"
```

```python
data = np.loadtxt(sn04gt, skiprows=1)
```

```python
data_lst = data.tolist()
data_lst
```

```python
wvl_lst = []
for row in data_lst:
    non_zero = 0
    for i in range(1, len(row)):
        if row[i] != 0:
            non_zero += 1
    if non_zero == 13:
        wvl_lst.append(row[0])
```
